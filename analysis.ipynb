# # 🏠 House Price Prediction - Exploratory Data Analysis (EDA)

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

plt.style.use("ggplot")
sns.set_style("whitegrid")
```

```python
# read the dataset
df = pd.read_csv("housing.csv")

df.head()
```

```python
# data information

print("Shape :", df.shape)

print("\nColumns")
print(df.columns)

print("\nData Types")
print(df.dtypes)
```

```python
# missing values
df.isnull().sum()
```

```python
# duplicate values
df.duplicated().sum()
```

```python
#statistical summary
df.describe().T
```

```python
# Histogram

#A histogram is used to understand the distribution of numerical features.

# Select numerical columns
numerical_columns = df.select_dtypes(include=np.number).columns

# Plot histogram
df[numerical_columns].hist(figsize=(18, 12), bins=30)

plt.suptitle("Distribution of Numerical Features", fontsize=20)
plt.tight_layout()

plt.show()
```

```python
# Box Plot :-Box plots help identify outliers in the dataset
plt.figure(figsize=(12,6))

sns.boxplot(data=df.select_dtypes(include=np.number))

plt.xticks(rotation=90)

plt.title("Box Plot of Numerical Features")

plt.show()
```

```python
#  Distribution of Target Variable
plt.figure(figsize=(8,5))

sns.histplot(df["median_house_value"], bins=40, kde=True)

plt.title("Distribution of House Prices")

plt.show()
```

```python
# Ocean Proximity Distribution
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="ocean_proximity",
    order=df["ocean_proximity"].value_counts().index
)

plt.xticks(rotation=45)

plt.title("Ocean Proximity")

plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))

corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.show()
```

# median income vs House price

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="median_income",
    y="median_house_value",
    alpha=0.5
)

plt.title("Median Income vs House Price")

plt.show()
```

# Latitude vs Longitude

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="longitude",
    y="latitude",
    hue="median_house_value",
    palette="viridis"
)

plt.title("House Locations")

plt.show()
```

# Pair Plot

```python
sns.pairplot(
    df[
        [
            "median_income",
            "housing_median_age",
            "total_rooms",
            "median_house_value"
        ]
    ]
)

plt.show()
```

```python
# Average House Price
average_price = df["median_house_value"].mean()

print(f"Average House Price : ${average_price:,.2f}")
```

```python
# Most Expensive Locations
top_locations = (
    df.groupby("ocean_proximity")["median_house_value"]
      .mean()
      .sort_values(ascending=False)
)

top_locations
```

```python
# plotting
plt.figure(figsize=(8,5))

top_locations.plot(kind="bar", color="skyblue")

plt.ylabel("Average House Price")
plt.title("Average House Price by Ocean Proximity")

plt.show()
```

# Highest Median Income Areas

```python
income = (
    df.groupby("ocean_proximity")["median_income"]
      .mean()
      .sort_values(ascending=False)
)

income
```

# Rooms vs House Price

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="total_rooms",
    y="median_house_value",
    alpha=0.4
)

plt.title("Total Rooms vs House Price")

plt.show()
```

# Population Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["population"], bins=40, kde=True)

plt.title("Population Distribution")

plt.show()
```

# Room Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["total_rooms"], bins=40, kde=True)

plt.title("Rooms Distribution")

plt.show()
```

# Bedrooms Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["total_bedrooms"], bins=40, kde=True)

plt.title("Bedrooms Distribution")

plt.show()
```

```python
# Business Insights

### Key Findings

# - Median income has the strongest positive relationship with house prices.
# - Houses located near the ocean are generally more expensive.
# - The dataset contains outliers in total rooms, total bedrooms, and population.
# - Missing values are mainly present in the total_bedrooms column.
# - House prices are not normally distributed.
# - Latitude and longitude significantly influence house prices due to geographical location.
# - Total rooms alone cannot accurately predict house prices.
```
